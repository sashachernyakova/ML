```python
import seaborn as sns
sns.set_theme(style="white")
sns.color_palette("Set1")
```

```python
# Эта строка создает объект PairGrid, который представляет собой сетку
# для визуализации парных отношений между различными признаками в DataFrame df.
# Параметр diag_sharey=False указывает, что графики на диагонали (для каждого признака отдельно) будут иметь разные шкалы по оси y. Параметр hue="target" указывает, что данные будут разделены по столбцу "target", и каждый уникальный класс будет отображен разным цветом.
g = sns.PairGrid(df, diag_sharey=False, hue="target")
# Эта строка отображает диаграммы рассеяния (scatter plots) в верхнем треугольнике
# сетки графиков. Для этого используется функция sns.scatterplot, а параметр s=15 указывает размер точек на графике.
g.map_upper(sns.scatterplot, s=15)
#  оценка плотности в отличие от гистограммы с использованием операции свертки
g.map_lower(sns.kdeplot)
# Параметр lw=2 устанавливает толщину линии для графиков
# график ядерной оценки плотности
g.map_diag(sns.kdeplot, lw=2)
g.add_legend()
```

```python
import numpy as np
```

```python
# names=True говорит о том, что первая строка файла содержит названия столбцов. dtype=None позволяет NumPy
# самостоятельно определить типы данных в столбцах.
data = np.genfromtxt('iris.csv', delimiter=',', skip_header=1,dtype=float, usecols=(1,2,3,4,5), filling_values=0)
data
```

```python
print("Первые 10 наблюдений:")
data[:10]
```

```python
# axis=0 означает агрегацию по столбцам (по вертикали)

# Рассчет характеристик с использованием методов NumPy
count = np.count_nonzero(~np.isnan(data), axis=0)
mean = np.mean(data, axis=0)  # Среднее значение
std = np.std(data, axis=0)    # Стандартное отклонение
min_value = np.min(data, axis=0)  # Минимальное значение
percentiles = np.percentile(data, [25, 50, 75], axis=0)  # 25-й, 50-й (медиана) и 75-й процентили
max_value = np.max(data, axis=0)  # Максимальное значение

feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'target']

# Вывод результатов
print("Feature names:", feature_names)
print("count:", count)
print("mean:", mean)
print("std:", std)
print("min:", min_value)
print("25%:", percentiles[0])
print("50%:", percentiles[1])
print("75%:", percentiles[2])
print("max:", max_value)
```

```python
# axis=1 - указывает, что мы удаляем столбец, а не строку. (axis=0 используется для удаления строк)
# inplace=True - это параметр, который указывает Pandas модифицировать DataFrame на месте, то есть без необходимости присваивать результат обратно переменной. Если inplace=False или не указано, то метод drop() вернет новый DataFrame с удаленным столбцом, не изменяя исходный DataFrame.
df = pd.read_csv('iris_modified.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)

numpy_arr = df[["sepal length (cm)", "sepal width (cm)",
"petal length (cm)", "petal width (cm)"]].to_numpy()
numpy_arr
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler

# Создание объектов для каждого из методов масштабирования
scalers = [StandardScaler(), MinMaxScaler(), MaxAbsScaler(), RobustScaler()]
scaler_names = ["StandardScaler", "MinMaxScaler", "MaxAbsScaler", "RobustScaler"]

# Применение каждого метода масштабирования к данным и построение гистограмм
fig, axs = plt.subplots(4, len(numpy_arr[0]), figsize=(20, 20))
features = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]

for i, scaler in enumerate(scalers):
    scaled_data = scaler.fit_transform(numpy_arr)
    for j in range(len(numpy_arr[0])):
      # В данном контексте [:, j] означает извлечение всех строк и только j-го столбца.
        axs[i, j].hist(scaled_data[:, j], bins=20)
        axs[i, j].set_title(f"{scaler_names[i]}: {features[j]}")


plt.show()
```

```python
def my_standard_scaler(X):
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    scaled_X = (X - mean) / std
    return scaled_X

scaled_numpy_arr = my_standard_scaler(numpy_arr)
scaled_numpy_arr
```

```python
# Рассчет характеристик с использованием методов NumPy

min_value = np.min(scaled_numpy_arr, axis=0)  # Минимальное значение
max_value = np.max(scaled_numpy_arr, axis=0)  # Максимальное значение
mean = np.mean(scaled_numpy_arr, axis=0)  # Среднее значение
variance_values = np.var(scaled_numpy_arr, axis=0)

feature_names = ['A', 'B', 'C', 'D', 'E']

# Вывод результатов
print("Feature names:", feature_names)
print("min:", min_value)
print("max:", max_value)
print("mean:", mean)
print("var:", variance_values)
```
